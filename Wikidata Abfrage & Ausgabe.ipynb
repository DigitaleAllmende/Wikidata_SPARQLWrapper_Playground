```python
# https://www.wikidata.org/wiki/User:ChristianSW/Kulturerbe#KulturLandschaftsElementeKataster

from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd

sparql = SPARQLWrapper("https://query.wikidata.org/sparql")

sparql.setQuery("""
SELECT DISTINCT ?item ?itemLabel ?pic ?coord ?klekslink WHERE {
  ?item (wdt:P31/wdt:P279*) wd:Q846659.
  ?item wdt:P5556 ?kleks.
  ?item wdt:P625 ?coord.
  OPTIONAL { ?item wdt:P18 ?pic. }
  BIND(URI(CONCAT("https://www.kleks.app/editor/?element_id=", STR(?kleks))) AS ?klekslink)
  SERVICE wikibase:label { bd:serviceParam wikibase:language "de,en". }
}
""")

sparql.setReturnFormat(JSON)
results = sparql.query().convert()

results_df = pd.json_normalize (results['results']['bindings'])
results_df[['item.value', 'itemLabel.value', 'coord.value', 'klekslink.value']].head()
```
